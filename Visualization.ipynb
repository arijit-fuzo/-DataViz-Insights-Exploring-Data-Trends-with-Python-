```python
import pandas as pd 
# reading the database 
data = pd.read_csv("Data.csv") 


# printing the top 10 rows 
display(data.head(10))
display(data.tail(10))
```

```python
# Scatter Plot
# Scatter Plot: Shows relationships between two continuous variables
import matplotlib.pyplot as plt 

# Scatter plot with day against tip 
plt.scatter(data['day'], data['tip']) 
# Adding Title to the Plot 
plt.title("Scatter Plot") 
# Setting the X and Y labels
plt.xlabel('Day') 
plt.ylabel('Tip') 
plt.show()
```

```python
# Bar Chart
# Bar Chart: Compares quantities of different categories
import matplotlib.pyplot as plt 

# Bar chart with day against tip 
plt.bar(data['day'], data['tip']) 
plt.title("Bar Chart")
# Setting the X and Y labels 
plt.xlabel('Day') 
plt.ylabel('Tip')
plt.show()
```

```python
# Histogram
# Histogram: Displays the distribution of a continuous variable by grouping data into bins

import matplotlib.pyplot as plt 

# histogram of total_bills 
plt.hist(data['total_bill']) 
plt.title("Histogram") 
# Adding the legends 
plt.show()
```
